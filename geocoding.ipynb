```python
# Import CSV to dataframe
df = pd.read_csv(concert_csv)
# optionally display the df
# df
```

```python
# Pull list of unique cities from df
cities = df['City'].unique()
# print(cities)
```

```python
# Iterate through df and send address column
# Request code from Postman that returns a json with the coordinate results for each location

base_url = "https://nominatim.openstreetmap.org/search?q='"
url_suffix = "'&format=json"
city_df = pd.DataFrame(columns=['City','X','Y''JSON_copy'])
for i in cities:
    url = base_url + i + url_suffix
    payload = {}
    headers = {
      'User-Agent': user_agent
    }
    response = requests.request("GET", url, headers=headers, data=payload)
    r_json = response.json()
    if r_json[0]["lon"]:
        x = r_json[0]["lon"]
    else:
        x = 0
    if r_json[0]["lat"]:
        y = r_json[0]["lat"]
    else:
        y = 0
    new_row = pd.DataFrame({'City':[i], 'X':[x], 'Y':[y], "JSON_copy":[r_json]})
    city_df = pd.concat([city_df, new_row])
# optionally display city_df
city_df
```

```python
# Join XY coordinates from City_df to df
join_df = df.set_index('City').join(city_df.set_index('City'))

# export to csv
join_df.to_csv(output_csv)
```
